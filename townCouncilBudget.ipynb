## Concepts

The following function shows the spendings of a given (by name) town council sorted by "codi pantalla". You can observe that there is a pattern. The more digits there are the deeper is the level (NIVELL).

```python
def concepte_despesa_ajuntament(data, ajuntament):
    tmp = data.loc[(data.TIPUS_PARTIDA == 'D') & (data.NOM_COMPLERT == ajuntament),['DESCRIPCIO', 'IMPORT', 'CODI_PANTALLA', 'NIVELL']]
    tmp[['CODI_PANTALLA']] = tmp[['CODI_PANTALLA']].astype(str)
    return tmp.sort_values(['CODI_PANTALLA'])
    
pd.set_option('display.max_rows', 500)
concepte_despesa_ajuntament(data_16_aj_no0, "Ajuntament d'Ogassa")
```

# Ingressos - despesa

```python
#data_16_aj_no0

def get_import_ajuntament(data, codi_ens, tipus):
    return np.sum(data.loc[(data.CODI_ENS == codi_ens) & (data.TIPUS_PARTIDA == tipus), ['IMPORT']].IMPORT)

def get_ingressos_despesa_per_ajuntament(data):
    codi_ens = np.unique(data.CODI_ENS)
    res = np.empty((codi_ens.shape[0],3))
    
    i = 0;
    for ens in codi_ens:
        ingres = get_import_ajuntament(data, ens, 'I')
        despesa = get_import_ajuntament(data, ens, 'D')
        res[i] = [ens, ingres, despesa]
        i = i +1
        
    return res
        
```

```python
ingres_despesa = get_ingressos_despesa_per_ajuntament(data_16_aj_no0)
```

### Ratio outcome / income
We can observe that the ratio is 1/2 most of the times. Probably meaning that the output is added twice. This confirms what we suspected in the **concepts** section

```python
rati_income_outcome = ingres_despesa[:, 1] / ingres_despesa[:,2]
rati_income_outcome
```

### Spending (outcome) by level and town council

```python
def sum_outcome_ajuntament_level(data, ajuntament, level):
    tmp = data.loc[(data.TIPUS_PARTIDA == 'D') & (data.NOM_COMPLERT == ajuntament) & (data.NIVELL == level),['IMPORT']]
    return np.sum(tmp)

def print_outcome_by_level_ajuntament(data, ajuntament):
    for i in range(1,5):
        print('Level ' , i, ': %06.0f' %sum_outcome_ajuntament_level(data, ajuntament , i))
print_outcome_by_level_ajuntament(data_16_aj_no0, "Ajuntament d'Ogassa")

```
